```python

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

%matplotlib inline
```

```python
df = pd.read_csv(r'C:\Users\ADMIN\Downloads\data.csv')
df
```

```python
fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'fixed acidity', data = df)
plt.show()
```

```python
fig = plt.figure(figsize = (10,6))
sns.barplot(x = 'quality', y = 'residual sugar', data = df)
plt.show()
```

```python
bins = (2, 6.5, 8)
group_names = ['bad', 'good']
df['quality'] = pd.cut(df['quality'], bins = bins, labels = group_names)
```

```python
df
```

```python
lq = LabelEncoder()
df['quality'] = lq.fit_transform(df['quality'])
df['quality'].value_counts()
```

```python
x = df.drop('quality',axis = 1)
y = df['quality']
```

```python
x_train, x_test,y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)

```

```python
rfc = RandomForestClassifier(n_estimators=200)
rfc.fit(x_train, y_train)
pred_rfc = rfc.predict(x_test)
```

```python
print("Accuaracy score =",accuracy_score(y_test, pred_rfc))
print(classification_report(y_test, pred_rfc))
```

```python
svc = SVC()
svc.fit(x_train, y_train)
pred_svc = svc.predict(x_test)
```

```python
print("Accuaracy score =",accuracy_score(y_test, pred_svc))
print(classification_report(y_test, pred_svc))
```

```python
param = {
    'C': [0.1,0.8,0.9,1,1.1,1.2,1.3,1.4,1.5],
    'kernel':['linear', 'rbf'],
    'gamma' :[0.1,0.8,0.9,1,1.1,1.2,1.3,1.4,1.5]
}
grid_svc = GridSearchCV(svc,param, cv=10, verbose=2)
```

```python
grid_svc.fit(x_train, y_train)
```

```python
pred = grid_svc.predict(x_test)
print("Accuaracy score =", accuracy_score(y_test, pred))
print(classification_report(y_test, pred))
```

```python
rfc_eval = cross_val_score(estimator = rfc, X = x_train, y = y_train, cv = 10, verbose=2)
rfc_eval.mean()
```
